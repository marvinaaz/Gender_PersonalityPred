```python
import os
import glob
import pandas as pd

folder_path = '/content/drive/MyDrive/MRP_w_Nancy/Final_Deliverable/4573018/elea 2/transcriptions/Transcription_txtFiles/'
all_txt_files = glob.glob(os.path.join(folder_path, '*.txt'))

# Store all transcripts
transcript_data = []

for file in all_txt_files:
    with open(file, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        for line in lines:
            parts = line.strip().split('\t')
            if len(parts) == 3:
                timestamp, speaker, utterance = parts
                transcript_data.append({
                    'file': os.path.basename(file),
                    'timestamp': timestamp,
                    'speaker': speaker,
                    'utterance': utterance
                })

df_transcripts = pd.DataFrame(transcript_data)
df_transcripts.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

gender_map = {1: 'Female', 2: 'Male'}
df['Gender'] = df['group'].map(gender_map)

# Define LIWC emotion and cognition columns
emotion_cols = ['posemo', 'negemo', 'anx', 'anger', 'sad']
cognition_cols = ['cogmech', 'insight', 'cause', 'discrep', 'tentat', 'certain']

# Create emotion and cognition composite scores
df['Emotion_Score'] = df[emotion_cols].sum(axis=1)
df['Cognition_Score'] = df[cognition_cols].sum(axis=1)

# Plotting Emotion vs Cognition scores by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Gender', y='Emotion_Score', palette='coolwarm')
sns.stripplot(data=df, x='Gender', y='Emotion_Score', color='black', alpha=0.4, jitter=True)
plt.title('Emotion Score Distribution by Gender')
plt.ylabel('Composite Emotion Score')
plt.xlabel('Gender')
plt.grid(axis='y')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Gender', y='Cognition_Score', palette='viridis')
sns.stripplot(data=df, x='Gender', y='Cognition_Score', color='black', alpha=0.4, jitter=True)
plt.title('Cognition Score Distribution by Gender')
plt.ylabel('Composite Cognition Score')
plt.xlabel('Gender')
plt.grid(axis='y')
plt.show()

```
